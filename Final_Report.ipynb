## VGG - Zeyu Zhang zz2941

#### Introduction on VGGNet

The full name of VGG is the Visual Geometry Group. It has released a series of convolutional network models beginning with VGG, which can be applied to face recognition and image classification. The original purpose of VGG's research on the depth of convolutional networks is to understand how the depth of convolutional networks affects the accuracy and accuracy of large-scale image classification and recognition. In order to deepen the number of network layers and to avoid too many parameters, a small 3x3 convolution kernel is used in all layers.

#### The network structure

The input of VGG is set to an RGB image of 224x244 size. The average RGB value is calculated for all images on the training set image, and then the image is input to the VGG convolutional network. A 3x3 or 1x1 filter is used, and the convolution step is fixed. In addition, the VGG network is not followed by a pooling layer behind each convolutional layer, or a total of 5 pooling layers distributed under different convolutional layers. VGG-16 mainly has three parts: Convolution, Pooling, and Fully Connected Layers.

Convolution layer: In this layer, filters are applied to extract features from images. The most important parameters are the size of the kernel and stride.

Pooling layer: Its function is to reduce the spatial size to reduce the number of parameters and computation.

Fully Connected Layers: These are fully connected connections to the previous layers as in a simple neural network.

The 16 in VGG16 refers to 16 layers that have weights. In VGG16 there are thirteen convolutional layers, five Max Pooling layers, and three Dense layers which sum up to 21 layers but it has only sixteen weight layers i.e., learnable parameters layer.

VGG16 takes input tensor size as 224, 244 with 3 RGB channel

Most unique thing about VGG16 is that instead of having a large number of hyper-parameters they focused on having convolution layers of 3x3 filter with stride 1 and always used the same padding and maxpool layer of 2x2 filter of stride 2.
The convolution and max pool layers are consistently arranged throughout the whole architecture.

Conv-1 Layer has 64 filters, Conv-2 has 128 filters, Conv-3 has 256 filters, Conv 4 and Conv 5 has 512 filters.

In VGGNet, each convolution layer contains 2 to 4 convolution operations. The size of the convolution kernel is 3 3, the convolution step size is 1, the pooling kernel is 2 * 2, and the step size is 2. The most obvious improvement of VGGNet is to reduce the size of the convolution kernel and increase the number of convolution layers.

## GoogLeNet - Yazhe Yan yy3177

### Introduction to GoogLeNet

GoogLeNet is a CNN deep learning method proposed by Google that has a lower error rate than the then VGGNet and won the 2014 ILSVRC. Unlike the previous idea that the main goal is to deepen the depth of the network, GoogLeNet proposes a new structure called inception.
 
Due to the fact that the location of the same item varies substantially across images, precise convolution processes are necessary for effectively identifying such objects. Large convolutional kernels often prioritize global information, whereas tiny convolutional kernels capture mostly local information. GoogLeNet uses many convolutional kernels of varying sizes at the same layer to gather input in order to tackle this challenge; this structure is called inception.
 
The most important idea of Inception is how the optimal local sparse structure in a convolutional network can be approximated and covered by easily available dense components. Convolutional layers are used in the construction of Inception networks; hence, the most essential goal is to identify the ideal local structure and then replicate it. The first step in the construction of a network using a layer-by-layer approach is to conduct an analysis of the correlation statistics of the layer immediately below it. Based on the results of this analysis, the highly correlated units are then divided into identical clusters, and these identical clusters are then connected to one another to form the units of the subsequent layer and to connect with the units of the layer below that. Additionally, in order to prevent problems with patch alignment, the filter sizes used are 1x1, 3x3, and 5x5. Additionally, pooling procedures are implemented across all of the layers.

The structure of inception is shown in the following figure.


image.png

Figure 8 the architecture of inception

Scan the layer below with three distinct convolution kernels while simultaneously executing pooling operations, and then concatenate the outputs of these four operations together to use as the input to the following layer. On the other hand, operations such as 3x3 or 5x5 will result in a filter with an output that is identical to the layer that was input. After concatenating the outputs of the pooling layer and the convolutional layer, there will be an excessive number of filters as a consequence. As a result, a more efficient structure is presented in order to cut down on the number of filters. The structure is shown in the following figure.

image.png

Figure 9 the architecture of improved inception

A 1x1 operation is added before the 3x3 and 5x5 convolutions, and a 1x1 operation is added after the 3x3 pooling operation. This structure may effectively minimize the total network size since the 1x1 convolution operation can reduce the number of filters. this approach of applying 1x1 convolution operation in Inception can make the total network broader. This conception is the fundamental structure of GoogLeNet. From the results, GoogLeNet employs the inception structure to construct a massive network with 22 layers, but with considerably lower parameters than earlier networks such as AlexNet.So the GoogleNet is a pretty great CNN structure.

## ResNet - Chang Gao cg3320

### Introduction on ResNet

Since the large-scale application of deep convolutional neural networks, in people's traditional cognition, in order to achieve higher accuracy, it is necessary to further increase the depth of the network. However, the emergence of the residual neural network (ResNet) has broken this perception. Based on Deep Residual Learning for Image Recognition from He et al, it is shown that there is a maximum threshold for the depth of traditional CNN models

image.png

Figure 14 the comparison of two training

In the previous practice, to solve a complex problem, we added some extra layers in the deep neural network, which improved the accuracy and performance. The intuition behind these operations is that more layers can learn more complex features. But now there is a threshold for the depth of the discovered layer. When the depth reaches a certain level, as the depth increases, the error rate also increases. As shown in the figure, compared to the 20-layer CNN, the 56-layer CNN has a significant increase in error rate. Vanishing gradient is the main cause of this problem, so the author introduced the concept of residual block to improve this problem.

image.png

Figure 15 Residual learning: a build block

Typical ResNet models are implemented with double- or triple- layer skips that contain nonlinearities (ReLU) and batch normalization in between. There are two main reasons for adding skip connections: to avoid the problem of gradient disappearance, and to avoid the problem of increased training error caused by too many layers.

## DenseNet

### Introduction on DenseNet

Dense Convolutional Network (DenseNet) is very similar to a ResNet with some-fundamental differences. ResNet is using an additive method that means they take a previous output as an input for a future layer, and in DenseNet takes all previous output as an input for a future layer as shown in the above image. With dense connection, fewer parameters and high accuracy are achieved compared with ResNet.
 
DenseNet connects each layer to other layers in a feed-forward fashion. In a traditional convolutional neural network, there are L connections for an L-layer network, while in DenseNet, there will be L(L+1)/2 connections. The input to each layer comes from the outputs of all previous layers. So in a dense net, we have less number of layers than the other model, so here we can train more than 100 layers of the model very easily by using this technique.


image.png

Figure 18 DenseNet structure

However, as we go deeper into the network, as the number of layers increases, the last layer accepts all previous layers as input and there will be a feature map explosion. In order to overcome this problem, we create a dense block here and so each dense block contains a prespecified number of layers inside them. And the output from that particular dense block is given to what is called a transition layer and this layer is like one-by-one convolution followed by Max pooling to reduce the size of
the feature maps. So, the transition layer allows for Max pooling, which typically leads to a reduction in the size of your feature maps.

image.png

Figure 19 DenseNet structure

So following are some Advantages of the dense net.
 
Strong Gradient Flow: The error signal can be easily propagated to earlier layers more directly. This is a kind of implicit deep supervision as earlier layers can get direct supervision from the final classification layer.
 
Parameter & Computational Efficiency: For each layer, the number of parameters DenseNet is much smaller than ResNet.
 
More Diversified Features: Since each layer in DenseNet receives all preceding layers as input, more diversified features and tends to have richer patterns.
 
Maintains Low Complexity Features: In Standard ConvNet, classifier uses most complex features. But in DenseNet, the classifier uses features of all complexity levels. It tends to give more smooth decision boundaries. It also explains why DenseNet performs well when training data is insufficient.

# Conclusion

For the four common cnn convolutional models, we conducted a cross-sectional comparison analysis from seven aspects, such as Appearance time, Number of layers, Top-5 Error, Data Augmentation, Inception, etc. The comparison results are as follows.

Through the introduction in the previous section and the comparison in the figure above, we can summarize the characteristics of the following four models：

VGGNet uses kernel factorization to decompose large convolutional kernels into small convolutional kernels, and further deepen the network to achieve better results. Among these, the kernel factorization approach is an excellent notion for reducing the parameters for the subsequent birth of GoogleNet. However, it only deepens the network without addressing the issue of gradient vanishing, making VGGNet extremely difficult to train.
GoogleNet increases the width of the convolutional neural network, convolves on multiple convolutional kernels of different sizes and then aggregates them, and uses 1*1 convolutional dimensionality reduction to reduce the number of parameters. This structure stacks the convolution (1×1, 3×3, 5×5) and pooling operations (3×3) commonly used in CNNs, which increases the width of the network on one hand and increases the adaptability of the network to scale on the other hand. Using 1×1 convolutional blocks (NIN) to reduce the number of features can reduce the computational burden of deep neural networks .


ResNet provides a dual-path identity mapping technique for the gradient disappearance and network degradation problems. ResNet addresses the problem of network model degradation by allowing the neural network to go deeper, allowing the network's ability to fit complex functions to be increased in a single step.

DenseNet takes the greatest aspects of ResNet and does additional inventive work on them, which enhances network performance even more. The dense connection can help to solve the gradient disappearance problem, improve feature propagation, stimulate feature reuse, and minimize the number of parameters significantly.
